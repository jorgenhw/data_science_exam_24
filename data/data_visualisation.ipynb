```python
import pandas as pd

# read data
climate_data = pd.read_csv('climate/AMOCdata.csv')
weather_data = pd.read_csv('weather/AarhusSydObservations/weatherAarhusSyd_Marts_april.csv')

climate_data = climate_data[:1250]
weather_data = weather_data[:1250]

weather_data['ds'] = pd.to_datetime(weather_data['DateTime'])
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# plot data
fig, ax = plt.subplots()
ax.plot(climate_data['time'], climate_data['AMOC0'])
ax.set_xlabel('Time')
ax.set_ylabel('Sea Temperature')
ax.set_title('Sea Temperature')
plt.show()

fig.savefig('figures/climate.png')

fig, ax = plt.subplots()
ax.plot(weather_data['ds'], weather_data['Middeltemperatur'])
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature in Aarhus')

# format x-axis
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()  # rotates the x-axis labels to fit them better

plt.show()

# save the plots
fig.savefig('figures/weather.png')
```
